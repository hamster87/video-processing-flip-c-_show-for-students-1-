```python
#用 imshow 會在圖框裡呈現, 畫面較小
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Image, clear_output

cap = cv2.VideoCapture("something.mp4") # Read the video

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            
    # Flip
    frame_flip_lr = 255-frame  ### left-right flip

    # Display the original and flipped frames
    combined_frame = np.hstack((frame, frame_flip_lr)) #horizontal_combine
   
    _,buffer =cv2.imencode(".jpg",combined_frame)
    img=Image(data=buffer.tobytes())
    

    clear_output(wait=True)
    display(img)
    
cap.release()
print("Processing complete.")
```
